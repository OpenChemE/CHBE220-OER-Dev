<h3> Common Units </h3>

We will go over four common unit types below, many other units are variants of these basic types of units.

<ul>
    <li> Mixer: has multiple feed streams and mixes these streams into one output stream </li>
    <p style="text-align:center;">
<img src="Lecture2-Images/Mixer.PNG" width=300>
</p>
    <li> Splitter: a splitter will split a feed stream into multiple streams with the same composition. The exiting streams are not necessarily equal in terms of flow rates, but will always have the same composition </li>
    <p style="text-align:center;">
<img src="Lecture2-Images/Splitter.PNG" width=300>
</p>
    <li> Separator: a separator is similar to a splitter, but has the ability to split streams into different exiting compositions </li>
     <p style="text-align:center;">
<img src="Lecture2-Images/Separator.PNG" width=300>
</p>
    <li> Reactor: a reactor typically consists of a vessel in which a reaction (or multiple reactions) take place. There are many different types of reactors, some of which you will learn more about in future courses (e.g. CHBE 376 & CHBE 355) </li>
    <p style="text-align:center;">
<img src="Lecture2-Images/Reactor.PNG" width=300>
</p>
    </ul>
    
 
Processes are often represented as interconnected process units consisting of mixers, reactors, splitters and separators among other unit operations.  
 
Examples of other common unit operations:
<ul>
    <li> Distillation (a separation process using the difference in boiling points of components)</li>
    <li> Absorption (removing components from a vapour stream using a liquid) and stripping (removing components from a liquid stream using a vapour) </li>
    <li> Extraction (a process that purifies a dissolved substance using its solubility in two immiscible liquids)</li>
    <li> Heat exchanger </li>
    <li> Filtration </li>
    </ul>

<h3>Example: Vinyl Chloride Production Process</h3>

Recall the vinyl chloride production process from the last class:
    
\begin{align*}
C_{2}H_{4} + Cl_{2} → C_{2}H_{4}Cl_{2}&\;\;\;\; (1)\\
C_{2}H_{4}Cl_{2} → C_{2}H_{3}Cl + HCl&\;\;\;\; (2)     
\end{align*}    
    
We now consider an additional reaction in the vinyl chloride production process. Most vinyl chloride manufacturers use the combination of direct chlorination and oxychlorination, which is called the <em> balanced process </em>. 
    
\begin{align*}
C_{2}H_{4} + Cl_{2} → C_{2}H_{4}Cl_{2}&\;\;\;\; (1)\;\;\;\; \\
C_{2}H_{4} + 2HCl + \frac{1}{2}O_{2} → C_{2}H_{4}Cl_{2} + H_{2}O&\;\;\;\; (2) \\
C_{2}H_{4}Cl_{2} → C_{2}H_{3}Cl + HCl&\;\;\;\; (3)
\end{align*}
    
<blockquote>    
<ol>
    <li> Direct Chlorination (last lecture) </li>
    <li> Oxychlorination </li>
    <li> Ethylene dichloride pyrolysis (last lecture) </li> 
    </ol></blockquote>
    
What is the overall reaction and what advantage does this have over what we saw last lecture?    
    
    
To get the overall reaction, we add all the reactions and cancel any compounds that appear in both the reactants and products of the overall reaction. In the separation process, HCl is recycled in back to the oxychlorination reactor if it is not depleted to ensure no HCl is released into the environment. Reaction (1) is balanced with the rest of the process in a way to provide the source of $C_{2}H_{4}Cl_{2}$ for reaction (3). This process gives us only 1 by-product: water. Using the balanced process gives us water as a by-product as opposed to hydrogen chloride. 
    
    
$$2C_{2}H_{4}+Cl_{2}+\frac{1}{2}O_{2} → 2C_{2}H_{4}Cl_{2} + H_{2}O $$
    
This process is carried out in multiple unit operations. To achieve the maxium effiency and avoid releasing hazardous chemicals, hydrogen chloride produced in reaction 3 is recycled back to the oxychlorination reactor and used reaction (2), as well as unreacted EDC(ethylene dichloride, $C_{2}H_{4}Cl_{2}$) in reaction 3 is recycled to react again.

<p style="text-align:center;">
<img src="Lecture2-Images/VCL1.PNG" width=600>
</p>
    
<blockquote>
    <ul>
        <li> Ethylene enters the process and is split into 2 streams: 1 stream goes to the oxychlorination reactor an the other stream goes to the direct chlorination reactor  </li>
        <li> Air and/or oxygen are fed into the oxychlorination reactor and chlorine is fed into the direct chlorination reactor </li>
        <li> EDC produced by both the oxychlorination and direct chlorination processes, and is fed into a purification unit to separate from the other components in the system </li>
        <li> EDC is then fed into the EDC pyrolysis reactor, followed by the VCM(vinyl chloride monomer, $C_{2}H_{3}Cl$) purification unit, which separates HCl, VCM, and EDC into 3 separate streams </li>
    </ul></blockquote>

<h2> Material Balances </h2> 

General material balance rules:
<ul>
    <li> What goes in must come out (or react) </li>
    <br>
    <li> Total mass into and out of a unit should always be equal </li>
    <br>
    <li> Molecules or moles may not be conserved, but the atoms should be </li></ul>
    
We won't cover material balances in depth in this course. Instead, I will rely on you applying the concepts you are learning in your Material and Energy Balances course.
