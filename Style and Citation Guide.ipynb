<h3>Example: Formation of Nitrogen Dioxide</h3>

Consider the formation of nitrogen dioxide 
$$ \frac{1}{2} N_{2} (g) + O_{2}(g) → NO_{2} (g)$$ $$\Delta H_{r} = 33.2 kJ$$

The formation of 1 mole of $NO_{2}$ yields an enthalpy change of 33.2 kJ
    
</div>

<div style="background-color:rgba(215,192, 75, 0.15)"> 
<h3>Exercise: Process Paths</h3>
<br>
What process paths can be taken to calculate the change in enthalpy for acetone going from $25^{\circ}C$ in the liquid phase to $60^{\circ}C$ in the vapour phase?

    
<br>    
    
</div>
    
    
<details>
  <summary><font color='blue'>Click here to see the solution!</font></summary>
<div style="background-color:rgba(0, 0, 0, 0.05)"> 
    
**Solution**
<br/>

**Step 1**: Bring the acetone to the boiling point temperature from $25^{\circ}C$ without changing the phase using the $C_{P}$ in the liquid phase.
<br/>
<br/>
**Step 2**: Use the latent heat (or heat of vaporization) to calculate the enthalpy of changing the phase from liquid to vapour.
<br/>
<br/>
**Step 3**: Bring the acetone to $60^{\circ}C$ from the boiling point temperature using the $C_{P}$ in the vapour phase.
    

</div>
</details>
