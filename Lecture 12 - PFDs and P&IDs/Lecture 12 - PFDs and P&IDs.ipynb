<h2> Process Flow Diagrams (PFDs) </h2> 

Process flow diagrams usually include:
<ul>
    <li> Major pieces of equipment that are represented by a description and a unique equipment number and name</li>
    <p style="text-align:center;">
    <img src="Lecture12-Images/PFD1.PNG" width=500>
    </p>
    <li> Process flow streams that are represented by a number and sometimes include process conditions and chemical composition of each stream. This information can be on the PFD itself or on an accompanying flow summary table</li>
    <p style="text-align:center;">
    <img src="Lecture12-Images/PFD2.PNG" width=500>
    </p>
    <li> Utility streams for major equipment</li>
    <p style="text-align:center;">
    <img src="Lecture12-Images/PFD3.PNG" width=600>
    </p>
    <li> Basic control loops that present the control strategy used to operate the processes </li>
    <li> An equipment summary table with key details</li>
</ul>
    
      
 <p style="text-align:center;">
    <img src="Lecture12-Images/PFDExp.jpg" width=1000>
    </p>

<h3> PFD Summary Tables </h3>

<h4> Stream Summary Table</h4>
The following essential information must be included in stream summary tables:
<ul>
    <li> Stream number</li>
    <li> Temperature</li>
    <li> Pressure</li>
    <li> Vapour fraction</li>
    <li> Total mass flowrate </li>
    <li> Total mole flowrate </li>
    <li> Individual component mole flowrates </li>
</ul> 
<br>
Other information may or may not be included as well:
<ul>
    <li> Stream name</li>
    <li> Component mole fractions </li>
    <li> Component mass fractions </li>
    <li> Individual component mass flowrates </li>
    <li> Volumetric flowrates </li>
    <li> Significant physical properties (density, viscosity, etc.)</li>
    <li> Thermodynamic data (heat capacity, specific enthalpy)</li>
</ul>

<h2> Piping and Instrumentations Diagrams (P&IDs)</h2>
<br>
P&IDs are based off of PFDs but include more specific information
<br>
Piping and instrumentation diagrams do not include:
<ul>
    <li> Operating conditions (T,P) </li>
    <li> Stream flows </li>
    <li> Equipment locations </li>
    <li> Supports, structures and foundations </li>
    <li> Pipe routings (lengths, fittings such as elbows, but connections above a certain size are generally shown) </li>
</ul>
