<h2> Process Flow Diagrams (PFDs) </h2> 
 

Process flow diagrams usually include:
<ul>
    <li> Major pieces of equipment that are represented by a description and a unique equipment number and name</li>
    <p style="text-align:center;">
    <img src="Lecture12-Images/PFD1.PNG" width=600>
    </p>
    <li> Process flow streams that are represented by a number and sometimes include process conditions and chemical composition of each stream. This information can be shown on the PFD itself or on an accompanying flow summary table</li>
    <p style="text-align:center;">
    <img src="Lecture12-Images/PFD2.PNG" width=600>
    </p>
    <li> Utility streams for major equipment</li>
    <p style="text-align:center;">
    <img src="Lecture12-Images/PFD3.PNG" width=600>
    </p>
    <li> Basic control loops that present the overarching control strategy used to operate the processes. These are generally optional, and not shown in the diagrams above. </li>
    <li> An equipment summary table with key details (shown later in this notebook)</li>
</ul>

<h3> PFD Equipment Numbering </h3>

This is a typical way to label a piece of equipment using its position in the process. This is not the only way to label equipment, but a very common way. We will stick with this in this class and you will likely see it very commonly.

<p style="text-align:center;">
<strong>P-101A</strong>
</p>

<ul>
    <li> First letter: equipment (P = pump)</li>
    <li> First number: plant section (1 = section 1 of plant)</li>
    <li> Last numbers: equipment number (01 = pump 1 in this section)</li>
    <li> Last letters: show duplicates/triplicates when two or more of the same equipment is used for the same stage of process. 
</ul>

<h4> Stream Summary Table</h4>
Certain essential information must be included in stream summary tables, this may vary slightly depending on the industry, what we will stick with in this class is the following:
<ul>
    <li> Stream number</li>
    <li> Temperature</li>
    <li> Pressure</li>
    <li> Vapour fraction</li>
    <li> Total mass flowrate </li>
    <li> Total mole flowrate </li>
    <li> Individual component mole flowrates </li>
</ul> 
<br>
Other information may or may not be included as well:
<ul>
    <li> Stream name</li>
    <li> Component mole fractions </li>
    <li> Component mass fractions </li>
    <li> Individual component mass flowrates </li>
    <li> Volumetric flowrates </li>
    <li> Significant physical properties (density, viscosity, etc.)</li>
    <li> Thermodynamic data (heat capacity, specific enthalpy)</li>
</ul>

<h2> Piping and Instrumentations Diagrams (P&IDs)</h2>
<br>
P&IDs are based off of PFDs but include more specific information
<br>
P&IDs do not include:
<ul>
    <li> Operating conditions (T,P) </li>
    <li> Stream flows </li>
    <li> Equipment locations </li>
    <li> Supports, structures and foundations </li>
    <li> Pipe routings (lengths, fittings such as elbows, but connections above a certain size are generally shown) </li>
</ul>

Instrument connections
<blockquote>     
<ul>
    <li>Capillary: fluid flow control</li>
    <li>Pneumatic: air or gas flow control </li>
    <li>Electrical: electrical currents controls </li>
</ul> </blockquote>


 <p style="text-align:center;">
    <img src="Lecture12-Images/InstrumentLoc.PNG" width=300>
    </p>

Instrument locations
<blockquote>     
<ul>  
    <li>Located in plant: generally means the instrument is near what it is measuring (local), it does not go to the control room</li>
    <li>Front panel in control room: Control rooms used to have instrument panels and would have manual controls on the front panel facing the operators, these have been replaced largely with digital (or virtual) screens which the operators can switch between different views. You could think of the main view in these digital screens as the front panel, with the most important controls or indicators shown. </li>
    <li>Back panel in control room: place for instruments that are perhaps less important or do not need to be looked at frequently. In our class we won't distinguish between these, but this is common notation you may see in P&IDs </li>
</ul> </blockquote>
