<h2> Control System Elements </h2>

Some common terms used in process control: 
<ul>
    <li><strong>Set-Point</strong>: this is a target or desired value of a variable in the system (such as flowrate, temperature, etc.) </li>
    <br>
    <li><strong>Disturbance Variable</strong>: a variable that we have no control over in the process but has direct effects on the process. </li>
    <br>
    <li><strong>Manipulated Variable</strong>: a variable that we can control in the process and directly affects the output of the process. </li>
    <br>
    <li><strong>Process Variable</strong>: the variable in the system or process that we desire to control. </li>
    <br>
    <li><strong>Controlled Variable</strong>: the output process variable we compare to the set-point.</li>
</ul>

<h3> Flow Measurement </h3>
<blockquote>
<strong>1 - Obstruction Meter</strong>
<p style="text-align:center;">
<img src="Lecture15-Images/ObstructionMeter.PNG" width=500>
</p>
An obstruction meter measures the change in pressure $\Delta P$ and correlates it to flow.    
<br>
    
<strong>2 - Rotational or Turbine Flowmeter </strong>
<p style="text-align:center;">
<img src="Lecture15-Images/TurbineMeter.PNG" width=500>
</p> 
A rotational or turbine flowmeter measures the speed of the turbine and correlates it to flow.
<br>
    
</blockquote>   

<p style="text-align:center;">
<img src="Lecture15-Images/FlowControl.PNG" width=500>
</p> 

For the flow control loop depicted above:
<ul>
    <li> If the flow is too low → open the valve </li>
    <li> If the flow is too high → close the valve </li>
</ul>

The transmitter is generally placed sufficiently upstream of the valve so that the flow reading is typically not affected by the valve.

<h3> Level Measurement </h3>
<br>
Many level measurement devices are available (pressure, weight, thermal), but float and sonic/radar level measures are the most common.

<blockquote>
<strong>1 - Float</strong>
<p style="text-align:center;">
<img src="Lecture15-Images/Float.PNG" width=500>
</p>   
<br>
    
<strong>2 - Sonic/Radar </strong>
<p style="text-align:center;">
<img src="Lecture15-Images/Sonic.PNG" width=450>
</p> 
A sonic or radar measure measures level of the tank by measuring the time (distance) it takes to reach the liquid-gas interphase through the waves reflected back to the device. 
<br>
    
</blockquote> 

A typical level control loop is depicted:
<p style="text-align:center;">
<img src="Lecture15-Images/LevelControl.PNG" width=550>
</p>

For the level control loop depicted above:
<ul>
    <li> If the level is too low → close the valve below the tank (outlet valve) </li>
    <li> If the level is too high → open the valve below the tank (outlet valve) </li>
</ul>

Note that we could also manipulate the inlet valve to affect the tank level, but typically in these systems the outlet valve is used to control the level. We will stick with this convention in this class.

<h3> Pressure Measurement </h3>

<blockquote>
<strong>1 - Manometer</strong>
<p style="text-align:center;">
<img src="Lecture15-Images/Manometer.PNG" width=200>
</p>   
<br>
Manometers measure difference in fluid height and correlate it to pressure.
    
<strong>2 - Differential Pressure Cells </strong>
<p style="text-align:center;">
<img src="Lecture15-Images/Diaphragm.PNG" width=450>
</p> 
A differential pressure cell measures the deflection of the membrane and correlates it to change in pressure $\Delta P$. 
<br>
    
</blockquote> 

A typical liquid pressure control loop is depicted:
<p style="text-align:center;">
<img src="Lecture15-Images/PressureControl.PNG" width=550>
</p>

For the level control loop depicted above:
<ul>
    <li> If the pressure is too low → open the valve </li>
    <li> If the pressure is too high → close the valve </li>
</ul>

For a centrifugal pump (least expensive and most common), the line can be closed without immediate damage to the pump. Other pumps (positive displacement pumps) will use other strategies.
