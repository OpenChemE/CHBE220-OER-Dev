<h2> Control System Elements </h2>

Some common terms used in process control: 
<ul>
    <li><strong>Set-Point</strong>: this is a target or desired value of a variable in the system (such as flowrate, temperature, etc.) </li>
    <br>
    <li><strong>Disturbance Variable</strong>: a variable that we have no control over in the process but has direct effects on the process. </li>
    <br>
    <li><strong>Manipulated Variable</strong>: a variable that we can control in the process and directly affects the output of the process. </li>
    <br>
    <li><strong>Process Variable</strong>: the variable in the system or process that we desire to control. </li>
    <br>
    <li><strong>Controlled Variable</strong>: the output process variable we compare to the set-point.</li>
</ul>

A sonic or radar measure measures level of the tank by measuring the time (distance) it takes to reach the liquid-gas interphase through the waves reflected back to the device. 
<br>

A typical level control loop is depicted:
<p style="text-align:center;">
<img src="Lecture15-Images/LevelControl.PNG" width=550>
</p>

For the level control loop depicted above:
<ul>
    <li> If the level is too low → close the valve below the tank (outlet valve) </li>
    <li> If the level is too high → open the valve below the tank (outlet valve) </li>
</ul>

Note that we could also manipulate the inlet valve to affect the tank level, but typically in these systems the outlet valve is used to control the level. We will stick with this convention in this class.

<h3> Temperature Measurement </h3>

<blockquote>
<strong>1 - Expansion-based</strong>
<p style="text-align:center;">
<img src="Lecture15-Images/Thermometer.png" width=200>
Image from <a href="https://commons.wikimedia.org/wiki/File:Thermometer_(PSF).png" title="via Wikimedia Commons">Pearson Scott Foresman</a> / Public domain
</p>   
<br>
Expansion-based thermometers measure fluid expansion and correlate it to temperature.
    
<strong>2 - Electrical </strong>
<p style="text-align:center;">
<img src="Lecture15-Images/Thermocouple.PNG" width=450>
Image from <a href="https://commons.wikimedia.org/wiki/File:Thermocouple_circuit_Ktype_including_voltmeter_temperature.svg" title="via Wikimedia Commons">Nanite</a> / CC0
</p> 
Thermocouples or resistance temperature devices measure changes in electrical properties and correlate them to temperature.
<br>
    
</blockquote> 

A typical slow-response endothermic temperature control loop is depicted:
<p style="text-align:center;">
<img src="Lecture15-Images/SlowResponseControl.PNG" width=550>
</p>
<blockquote> A slow-response endothermic temperature control loop is classified as <strong>slow</strong> because it takes a longer time to manipulate steam to see a temperature change. When the steam is being manipulated, the response of the system is dependant on the amount of steam in the heat exchanger, which is delayed by the steam temperature change lag.</blockquote>
   


A typical critical endothermic temperature control loop is depicted:
<p style="text-align:center;">
<img src="Lecture15-Images/CriticalControl.PNG" width=550>
</p>
<blockquote> A critical endothermic temperature control loop is <strong>faster</strong> than a slow-response endothermic control loop because the temperature of the process fluid is directly changed. This eliminates the lag that would be present from manipulating the steam temperature. </blockquote>

<h3>Example: Endothermic Temperature Control Loops</h3>

Consider a heat exchanger where an endothermic temperature control loop is present in the system. There is a $5^{\circ}C$ (from the set-point) increase in the outlet temperature of the process fluid. Which endothermic control loop will decrease the temperature back to its original set-point faster and why?
    
For a <strong> slow-response endothermic temperature control loop </strong>:
<blockquote> When the temperature change is detected, the controller will open the valve slightly to allow for less hot fluid to flow through the heat exchanger. This will eventually cause the process fluid out temperature to decrease to the set-point temperature </blockquote>
    
For a <strong> critical endothermic temperature control loop </strong>: 
<blockquote> When the temperature change is detected, the controller will open the valve of the cold process fluid. This will directly add cold process fluid to the outlet process fluid and directly decrease the temperature back to its set-point, without any delay. </blockquote>
    
Therefore, a critical endothermic temperature control loop is going to decrease the temperature back to its original set-point faster.    
    
    
</div>
