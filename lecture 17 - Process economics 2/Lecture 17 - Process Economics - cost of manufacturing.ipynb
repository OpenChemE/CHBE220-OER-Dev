## Manufacturing Costs

Typically divided into three main elements:
1. <strong>Direct manufacturing costs</strong>:  Factors that vary directly with the rate of production
e.g. raw materials, plant operators

2. <strong>Fixed manufacturing costs</strong>: Factors not affected by the level of production, but related to manufacturing
e.g. taxes and insurance

3. <strong>General expenses</strong>: Management and administrative expenses not related to manufacturing

<font color="red"> $$ \text{Cost of Manufacturing }(COM)=\text{direct cost} + \text{fixed cost} + \text{general expenses} $$ </font>

## Number of Operating Labour

<strong> $N_{OL}$ - Number of operators needed in the plant at a given time to ensure it runs smoothly. </strong>

Number of operators ($N_{OL}$) needed per shift can be estimated from a correlation by Alkhayat and Gerrard* which is based on data from five major chemical companies

<font color="red">$$N_{OL}=(6.29+31.7P^2+0.23N_{np})^{0.5}$$</font>

where
> $N_{OL}$ - number of operating labour workers <strong>at a given time</strong>
>
> $P$ - process steps involving handling <strong> solids </strong> (transport, partical size control, particulate removal)
>
> $N_{np}$ - number of process steps <strong>not invloving handling solids</strong>, including compression, heating, cooling, mixing, separation and reaction, but <font color="red"> not including pumps and vessels </font>

## Cost of operating labour

<font color="red">$$\text{# of operator}=N_{OL}×\frac{\text{# of total shifts in a year}}{\text{# of shifts per operator}}$$</font>

The number of operator is <font color="red">rounded up to the nearest integer</font> when the calculation does not result in a whole number.

<font color="red">$$\text{cost of operating labour}=\text{# of operator}×\text{salary}$$</font>

The number of total shifts in a year and number of shifts per operator can be calculated as follows:

On average an operator works five eight-hour shifts per week for 49 weeks per year (3 weeks off for sick leave and vacation deducted from 52 total weeks in a year).

$$\text{# of shifts per operator} = 5 \frac{shifts}{week}×49\frac{weeks}{year}=245\text{ shifts}$$

A plant generally requires operators 365 days a year including maintenance and shutdowns. One day is split into 3 eight-hour shifts

$$\text{# of total shifts in a year}= 365 days× 3 \frac{shifts}{day}=1095 \text{ operating shifts}$$

The number of operators needed to fill a position that needs one operator at a given time is $\frac{1095}{245}≈4.5$.

To find the total cost of operating labour,  we can use an average salary for a plant operator (these statistics can generally be found from an agency such as Statistics Canada). 

Using the previous example ($N_{OL}=4$), we can calculate the total cost of operating labour:

total number of operators needed: $4×4.5=18$ The number should be rounded up if it does not result in a whole number.

The average salary for a plant operator can be found from an agency such as  <a href="https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=1410006401">Statistics Canada</a>. Say an operator receives a salary of 70,000 dollars per year, the total cost of operating labour is $18×70,000=1,260,000$ dollars per year

Fixed Costs of Manufacturing are costs that are not directly impacted by the manufacturing rate. To assess the impacts of some of the main elements of the fixed costs we will use multiplication factors based on a certain known costs.

## Depreciation

Governments will not let companies write off all expenses to build a plant immediately, instead they are written off over time in what is known as depreciation.

Depreciation is calculated based on the purchase value of equipment minus the salvage value and evaluated over a number of years 
specified. This does not include land costs, since land generally retains its value, whereas equipment and buildings tend to lose their value over time. 

Straight-line depreciation is the simplest calculation of depreciation:
<font color ="red">$$ \text{Depreciation}=\frac{\text{purchased value } (FCIL)-\text{salvage value}}{\text{life of equipment } (n)}$$ </font>

## Overall Manufacturing Costs

Summing together the Direct Costs of Manufacturing, Fixed Costs of Manufacturing and General Expenses, we can get a general formula to estimate the total cost of manufacturing.

Direct Cost of Manufacturing: $C_{RM}+C_{WT}+C_{UT}+1.33C_{OL}+0.03COM +0.069FCI$

Fixed Cost of Manufacturing: $0.708C_{OL}+0.068FCI+depreciation(0.1FCI)$

General Expenses: $0.177C_{OL}+0.009FCI+0.16COM$

<font color="red">$$\text{cost of manufacturing}=0.180FCI+1.23(C_{RM}+C_{WT}+C_{UT})+2.73C_{OL}+\text{depreciation}(0.1FCI)$$</font>

Depreciation is not included in most of the calculations we perform - depreciation is not a real cost that we need to pay, but it is part of the the money we lose for tax purposes.
