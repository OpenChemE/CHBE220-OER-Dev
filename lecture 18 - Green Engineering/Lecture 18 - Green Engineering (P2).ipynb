<h2> Life Cycle Assessment </h2>

<h3>Terminology</h3>
<ul>
    <li><strong>Cradle-to-grave:</strong> this scope includes all aspects of a product, starting from raw material extraction to the product's end of life (use and disposal) </li><br>
    <li><strong>Cradle-to-gate:</strong> this scope includes all aspects of a product from raw material extraction to production, but does not include the product's use or disposal </li><br>
    <li><strong>Gate-to-gate:</strong> this scope includes the manufacturing/production and processing aspects of a product only</li></ul>




Life Cycle Assessment (LCA) is a technique used to quantify the environmental impact of a product from raw material acquisition through end of life disposition (<strong>cradle-to-grave</strong>)

There are four stages of a life cycle assessment:
<ol> 
    <li><strong> Step 1: Goal Definition and Scoping: </strong> Identify the LCA's purpose, the products of the study, and determine the study boundaries. This can include things like what is included and what is not included in the study </li>
    <br>
    <li><strong>Step 2: Life-cycle Inventory:</strong> Quantify the energy and raw material inputs and environmental releases associated with each life cycle phase. </li>
    <br>
    <li><strong>Step 3: Impact Analysis:</strong> Assess the impacts on human health and the environment. </li>
    <br>
    <li><strong>Step 4: Interpret Results:</strong> Evaluate opportunities to reduce energy, material inputs, or environmental impacts at each stage of the product life-cycle.</li>
    </ol>
    
Life Cycle Assessments are based on international standards set by the International Organization for Standardization (ISO 14040). 
<blockquote>
Each of the steps are set by different ISOs: 
<ul>
    <li>Step 1: ISO 14041</li>
    <li>Step 2: ISO 14041</li>
    <li>Step 3: ISO 14042</li>
    <li>Step 4: ISO 14043</li>
    </ul></blockquote>

<h3> Step 1: Goal Definition and Scoping </h3>

Define the goal:
<ul>
    <li> What is the intended application of the study?</li>
    <li> Who is the intended audience?</li>
    </ul>
    
Define the scope:
<ul>
    <li> What is the product system to be studied?</li>
    <li> What is the functional unit?*</li>
    <li> What are the boundaries of the product system? </li>
    <li> What are the assumptions and limitations of the study? </li>
    <li> What impact categories need to be included in the study? </li>
    </ul>
    
*<strong>Functional Units</strong>: a measure of the function of the studied system. It provides a reference to which the inputs and outputs can be related and enables a comparison of two essentially different systems. 
<blockquote>
    Examples of functional units:
    <ul>
        <li> For a paint system (the paint as well as layers of coating materials applied to a surface), the functional unit may be defined as the unit surface protected for 10 years </li>
        <li> For a printer, the funcional unit may be defined as the number of printed pages of an acceptable print quality </li>
        <li> For a power generation system, the functional unit may be defined as 1kWh of electricity</li>
    </ul>
    </blockquote>

<h3> Example: Reusable Shopping Bags </h3>

Let's say you own a local grocery store in Kitsilano. Your customers start requesting that you sell reusable shopping bags to minimize waste. You think its a great idea, but want to make sure that the reusable shopping bags you sell have the lowest environmental impact. Since you are an engineer and learned about LCA in CHBE 220, you decide to perform a streamlined LCA. 
    
What is your <strong>goal</strong>?
<blockquote>
    Determine which shopping bag has the lowest environmental impact </blockquote>
    
After doing a search of what is available on the market, you determine that your options for shopping bags are as follows:
<ul> 
    <li> Single use paper </li>
    <li> Single use plasic </li>
    <li> Reusable plastic </li>
    <li> Reusable cotton </li>
    </ul>
    
What are your <strong> system boundaries and process</strong>?

    <li> Half of paper bags are recycled at end of life and the other half go to a landfill </li>
    <li> Plastic and cotton bags go to landfill at the end of life </li>
    </ul>

<h3> Step 2: Life Cycle Inventory </h3>

The life cycle inventory step is a highly data intensive step that is based on detailed material and energy balance analyses over the life cycle of a product. The aim of this step is to account quantitatively for the components which make up the life cycle for our products.

<strong>Advantages</strong>:
<ul>
    <li> Quantifiable data for each product to assess impacts (although these can be highly dependent on the assumptions we make in step 1) </li></ul>
    
<strong>Challenges</strong>:
<ul>
    <li> Assumptions made when data unavailable</li>
    <li> Can be very time intensive to account for all materials, energy and steps to create a product</li></ul>

<h2> Step 3: Impact Assessment</h2>

<strong>Impact Assessment Results</strong>: impact assessment converts the inventory into impact categories or end points, detailing human health and environmental impact.

<ul>
    <li> Converts the inventory into impact categories or mid/end points that characterize the environmental effects </li>
    <li> Impact categories may include: carcinogens, climate change, respiratory organics and inorganics, radiation, ozone layer, ecotoxicity, acidification/eutrophication, land use, minerals, fossil fuels </li>
    <li> We can add weights to impact categories to indicate if we want to weight them more or less significantly in an overall analysis (this is again where some assumptions may come in and so we have to be careful when setting up and interpreting LCAs) </li>

<p style="text-align:center;">
<img src="Lecture18-Images/Diagram4.PNG" width=800>
</p>  
    
The following figure shows an example of life-cycle assessment results for EV batteries from a study conducted by the United States Environmental Protection Agency (US EPA).[1]
<p style="text-align:center;">
<img src="Lecture18-Images/ImpactAss.PNG" width=600>
</p>
        

    
*ADP: abiotic depletion potential - depletion of non-living resources such as minerals, clay, fossil fuels, etc.<br>
    
*AP: acidification potential - the potential of the battery releasing harmful acids into the environment<br>
    
*EcoTP: ecological toxicity potential - the potential of the battery being toxic to surrounding ecology <br>
    
*EP: eutrophication potential - nutrient addition that leads to the (over)growth of aquatic plant species; can lead to things like algal blooms and food chain disruptions<br>
    
*GWP: global warming potential - the potential of warming caused by the release of greenhouse gases as caused by the battery<br>
    
*HTP: human toxicity potential - the potential the battery process has for releasing chemicals that are toxic (dosage is taken into account) into the environmental<br>
    
*OCH: occupational cancer hazard - the potential the battery process has for exposing people that work on this battery to carcinogens<br>
    
*ODP: ozone depletion potential - the potential the battery process has for releasing chemicals that degrade the ozone layer<br>
    
*OnCH: occupational non-cancer hazard - the potential the battery process has for exposing people that work on this battery to other diseases, injuries, or other health problems<br>
    
*POP: photochemical oxidation potential - the potential the battery process has for releasing chemicals that react in the sunlight and release harmful products into the environment<br>    
    
*EOL: end-of-life; note that the negative values for the EOL bar signify recovery of the material at the end of battery's life, which reduced the life-cycle impact.

<h2> Step 4: Report Results </h2>

<strong> Life cycle interpretation </strong>: findings of the inventory analysis or impact assessment are evaluated in relation to the  goal and scope of the study to reach conclusions and recommendations.

In this stage we aim to:
<ol>
    <li> Identify significant issues</li>
    <li> Evaluate results for completeness, consistency, and sensitivity of the data</li>
    <li> Draw conclusions and make recommendations consistent with the goal and scope of the study </li> </ol>
