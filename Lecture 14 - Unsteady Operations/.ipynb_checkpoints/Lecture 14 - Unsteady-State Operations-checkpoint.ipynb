<h2> Introduction to Unsteady-State Balances </h2>

<strong>Unsteady-State</strong>: in unsteady-state processes, mass and energy in the system is changing with time (not constant). 

Understanding and modeling these systems allow us to:
<ul>
    <li>Control system variables by adjusting parameters</li>
    <br>
    <li>Predict the system's behavior during start-up and shutdown</li>
</ul>

<h2> General Balance Equation </h2>

To characterize systems (generally process units), we can do a balance on any quantity (total mass, energy, moles, etc.). For this class we will focus on mass, energy and mole balances, but we could do these balances on any quantity.

$$Input - Output + Generation - Consumption = Accumulation$$

<blockquote>
    <ul>
        <li> Input: Enters through the system boundaries ($IN$) </li>
        <li> Output: Exits through the system boundaries ($OUT$)</li>
        <li> Generation: Produced in the system ($GEN$) </li>
        <li> Consumption: Consumed in the system ($CON$) </li>
        <li> Accumulation: Build-up or overall loss in the system ($ACC$) </li>
    </ul>
    </blockquote>
    
<em> Note that the generation and consumption terms are often associated with chemical reactions </em>

$$ IN - OUT + GEN - CON = ACC$$

Consider a reactor with continuous inlet and exit streams. A reaction occurs in the reactor that is first order:

$$ A → B $$

The inlet stream only contains species A and the reaction goes to completion in the reactor. The reactor is initially filled entirely with A. What terms in the following expression are considered (i.e. non-zero) for this system when assessing the moles of species A as the reaction starts occuring?
    
$$ IN - OUT + GEN - CON = ACC$$

    
<br>    
    
</div>

<br>

<h2> Transient balances </h2>

<strong>Transient balances</strong> are any balances that where quantities change with time. If we have a continuous process, we can analyze what happens to a given quantity for a given change in time $\Delta t$. As an example, let's take mass of a given species (we'll call species A) to be our quantity of interest:

$$ IN - OUT + GEN - CON = ACC$$ <br>
$$ \dot{m}_{in,A}*\Delta t - \dot{m}_{out,A}*\Delta t + \dot{r}_{gen,A}*\Delta t - \dot{r}_{con,A}*\Delta t = \Delta M_{A} $$

<blockquote>
    Where: 
    <ul>
        <li> $\dot{m}_{in,A}$, $\dot{m}_{out,A}$ are the mass flows in and out of the system of species A </li><br>
        <li> $\dot{r}_{gen,A}$, $\dot{r}_{con,A}$ are the rates of generation and consumption of A in the system</li><br>
        <li> $M_{A}$ is the total mass of A in the system at a given time</li>
    </ul>
</blockquote>
    
    
If we want to turn this into a differential balance, we will isolate $\Delta t$ and let it approach to 0 (just like in calculus classes):

$$ \dot{m}_{in,A}*\Delta t - \dot{m}_{out,A}*\Delta t + \dot{r}_{gen,A}*\Delta t - \dot{r}_{con,A}*\Delta t = \Delta M_{A} $$
<br>
$$ (\dot{m}_{in,A} - \dot{m}_{out,A} + \dot{r}_{gen,A} - \dot{r}_{con,A})*\Delta t = \Delta M_{A} $$
<br>
$$ (\dot{m}_{in,A} - \dot{m}_{out,A} + \dot{r}_{gen,A} - \dot{r}_{con,A}) = \frac{\Delta M_{A}}{\Delta t} $$
<br>
$$ (\dot{m}_{in,A} - \dot{m}_{out,A} + \dot{r}_{gen,A} - \dot{r}_{con,A}) = \frac{dM_{A}}{dt} $$
<br>

For steady-state processes, the $\frac{dM_{A}}{dt}$ term goes to 0, as there would be no change in the total mass of A over time. For unsteady-state processes, this differential equation must be solved using <strong>boundary conditions</strong>. 


<blockquote>
An example of a boundary condition is $M_{A}(t=0)= M_i$ <br>

This equations says that the mass of A in the system at time (t = 0) is equal to some mass ($M_i$). This mass can be equal to 0, meaning that there is no A in our system at time zero. It can also be equal to a specific value, meaning that our system starts with an initial mass of A. 
    
</blockquote>
    
    
    

<div style="background-color:rgba(75, 171, 215, 0.15)"> 
<h3>Example: Chemical Reactor</h3>
    
Consider a "continuous stirred-tank reactor" (CSTR). CSTRs are reactors with continuous feed and exit streams and some kind of mixer. Let's say the following first order reaction takes place in our CSTR: 
    
$$ A → B $$ <br>
    
Say the units of the rate of consumption of A are $\frac{mol}{s*L}$.
    
Say we know the following information about this CSTR: 
    
<ul>
    <li> The feed enters at a volumetric flowrate of $\dot{V}_{0}$ in L/s and a concentration of $C_{A,0}$ in mol/L </li>
    <br>
    <li> The volume of the tank is $V$ in L and it is perfectly mixed. This means that the concentration of A is uniform within the tank and matches the exit concentration. </li>
    <br>
    <li> Initially ($t=0$), the tank is filled to $V_i$ in L and has an initial concentration of $C_{A,i}$ in mol/L </li>
    <br>
    <li> The exit stream flows at a rate of $\dot{V}$ in L/s </li>
    <br>
    <li> We can assume that the density of all streams in the system is constant at $\rho$ in g/L </li>
</ul>

<p style="text-align:center;">
<img src="Lecture14-Images/cstr.PNG" width=400>
</p>    
    
We want to write a balance for the total mass in the system under transient conditions. 

    
We start off by writing out the transient mass balance that we derived earlier:     
$$ (\dot{m}_{in,A} - \dot{m}_{out,A} + \dot{r}_{gen,A} - \dot{r}_{con,A}) = \frac{dM_{A}}{dt} $$    <br>
    
Using separation of variables (from calculus), we can integrate both sides. We want to find a given value about our system at a specific final time ($t_{f}$), starting from an initial time (say $t_{0} = 0$)
    
$$ (\dot{m}_{in,A} - \dot{m}_{out,A} + \dot{r}_{gen,A} - \dot{r}_{con,A})dt = dM_{A} $$    
<br>
$$ \int^{t_{f}}_{t_{0}}(\dot{m}_{in,A} - \dot{m}_{out,A} + \dot{r}_{gen,A} - \dot{r}_{con,A})dt = \int^{M_{A,t_{f}}}_{M_{A,t_{0}}}dM_{A} $$
<br>
$$ \int^{t_{f}}_{t_{0}}(\dot{m}_{in,A} - \dot{m}_{out,A} + \dot{r}_{gen,A} - \dot{r}_{con,A})dt = M_{A,t_{f}}-{M_{A,t_{0}}} $$    
    
The integral expression above can be used to describe any system. The terms will change based on the problem.  
    
In this scenario, there is no mass generated or consumed. Therefore, we get the following mass balance:
    
$$ \int^{t_{f}}_{t_{0}}(\dot{m}_{in,A} - \dot{m}_{out,A})dt = M_{A,t_{f}}-{M_{A,t_{0}}} $$     
    
    
    
</div>    

<h3> Differential Balances General Procedure </h3>

<ol>
    <li> Omit any terms that are equal to zero or neglible from the mass, energy, or mole balances. </li>
    <br>
    <li> Write an expression for the total amount of the quantity of interest (mass, energy, moles) using the remaining terms in the balance </li>
    <br>
    <li> Determine the initial conditions of the system and solve accordingly </li>

<div style="background-color:rgba(215,192, 75, 0.15)"> 
<h3>Exercise: Transient Mass Balance</h3>
<br>
Consider pumping liquid from a tank that has one drain line as depicted below:
    
<p style="text-align:center;">
<img src="Lecture14-Images/DrainingTank.PNG" width=300>
</p>   

Assume the tank contains only 1 species (A) and density is constant. What are the relevant terms in the mass balance for this draining tank? 
    
Say at time zero, the height of the tank is 5 $m$, the tank also has a diameter of 2 $m$ and the amount of A being pumped out is 0.1 $m^3/min$. What is the height of the tank at 10 minutes? 
      
    
</div>
    
    
<details>
  <summary><font color='blue'>Click here to see the solution!</font></summary>
<div style="background-color:rgba(0, 0, 0, 0.05)"> 
    
<strong>Solution</strong>
<br/>

<strong>Step 1</strong>: Determine what terms in the general balance are zero or negligible.
<br/>
$$ IN - OUT + GEN - CON = ACC$$    
<br/>
Since there are no feed streams and no reactions (no mass generated or consumed), the $IN$, $GEN$, and $CON$ terms can be ommited from the balance. 
<br/>
$$ -OUT = ACC$$
<br>    
<strong>Step 2</strong>: Write out the mass balance for species A in the system.
<br/>
 $$ (-\dot{m}_{out,A}) = \frac{dM_{A}}{dt} $$   
<br/>
<strong>Step 3</strong>: Expand the mass terms.
<br>
Recall that $$Mass=Volume*density$$
<br>
$$(-\dot{V}_{out,A}*\rho) = \rho*\frac{dV_{A}}{dt}$$
<br>
Since the density is constant, the density terms on both sides cancel each other out
<br>
$$(-\dot{V}_{out,A}) = \frac{dV_{A}}{dt}$$
<br>    
The volume terms can be further expanded. Recall that the volume of a cylinder is $Area*Height$ or $A*h$. In differential terms, $\frac{dV}{dt}=A\frac{dh}{dt} + h\frac{dA}{dt}$. Since the area is constant in this case, this simplifies to: $\frac{dV}{dt} = A\frac{dh}{dt}$.
<br>
$$(-\dot{V}_{out,A}) = A_{tank}*\frac{dh}{dt}$$
<br>
$$A_{tank} = \frac{\pi}{4}*{D_{tank}}^{2}=\frac{\pi}{4}*{2 m}^{2}=3.14 m^{2}$$
<br>    
<strong>Step 4</strong>: Solve the integral from t = 0 to t = 10 mins    
<br>    
$$(-\dot{V}_{out,A})dt = A_{tank}*dh$$
<br>
$$\int^{t=10 min}_{0 min}(-\dot{V}_{out,A})dt =\int^{h_{final}}_{h_{intial}} A_{tank}*dh$$
<br>
$$\int^{t=10 min}_{0 min}(-0.10\frac{m^3}{min})dt =\int^{h_{final}}_{h_{intial}}3.14m^{2}*dh$$
<br>
$$(-0.1\frac{m^{3}}{min})*(10min-0min) = (3.14m^{2})*(h_{final}-5m)$$
<br>
$$-1 m^{3}*\frac{1}{3.14m^{2}} = h_{final}-5m$$
<br>
$$-0.32 m = h_{final}-5m$$
<br>    
$$(-.32+5) m = h_{final}$$
<br>
$$h_{final}=4.68 m$$    
    
</div>
</details>

<h3>Exercise: Transient Mole Balance</h3>
<br>

Consider a CSTR where carbonic acid irreversably decomposes to form water and carbon dioxide as follows:
    
$$ H_{2}CO_{3} → H_{2}O + CO_{2} $$
<p style="text-align:center;">
<img src="Lecture14-Images/Ex3.PNG" width=300>
</p>  
There are two streams entering the CSTR that both contain carbonic acid. The first stream has a concentration of $C_{H_{2}CO_{3},01}$ and a volumetric flowrate of $\dot{V}_{01}$, while the second stream has a concentration of $C_{H_{2}CO_{3},02}$ and a volumetric flowrate of $\dot{V}_{02}$. The volume of the reactor is $V$. The concentration of the exit stream is $C_{H_{2}CO_{3},i}$ at 0 minutes and is $C_{H_{2}CO_{3},f}$ after 40 minutes. If the reaction is first order with the following rate of reaction: 

$$-r_{H_{2}CO_{3}} = k*C_{H_{2}CO_{3}}$$
    
What is the over all mole balance?     
    
 
    
</div>
    
    
<details>
  <summary><font color='blue'>Click here to see the solution!</font></summary>
<div style="background-color:rgba(0, 0, 0, 0.05)"> 
    
<strong>Solution</strong>
<br/>
<strong>Step 1</strong>: Determine what terms are present in the mole balance.
<br> 
$$IN = \dot{V}_{01}*C_{H_{2}CO_{3},01} + \dot{V}_{02}*C_{H_{2}CO_{3},02}$$
<br>
$$OUT= \dot{V}*C_{H_{2}CO_{3}}$$
<br>
$$ GEN = 0$$
<br>
$$ CON = -r_{H_{2}CO_{3}}*V = k*C_{H_{2}CO_{3}}*V $$
<br>
$$ACC = \frac{dN_{H_{2}CO_{3}}}{dt} = \frac{d(V*C_{H_{2}CO_{3}})}{dt}$$
<br>
    
<strong>Step 2</strong>: Write out the mole balance with the derived terms.
<br>
    
$$ \dot{V}_{01}*C_{H_{2}CO_{3},01} + \dot{V}_{02}*C_{H_{2}CO_{3},02} - \dot{V}*C_{H_{2}CO_{3}} - k*C_{H_{2}CO_{3}}*V = \frac{d(V*C_{H_{2}CO_{3}})}{dt}$$     
    
<br>

<strong>Step 3</strong>: Solve the different equation using the intial conditions $C_{H_{2}CO_{3}}(t=40 min) = 0 M$ and  $C_{H_{2}CO_{3}}(t=0 min) = 0.5 M$
<br>
$$\int^{t=40mins}_{t=0mins}(\dot{V}_{01}*C_{H_{2}CO_{3},01} + \dot{V}_{02}*C_{H_{2}CO_{3},02}-\dot{V}*C_{H_{2}CO_{3}}-k*C_{H_{2}CO_{3}}*V)dt=\int^{C_{H_{2}CO_{3}}(t=40 min)}_{C_{H_{2}CO_{3}}(t=0 min)}V*d(C_{H_{2}CO_{3}})$$
<br>
$$(\dot{V}_{01}*C_{H_{2}CO_{3},01} + \dot{V}_{02}*C_{H_{2}CO_{3},02}-\dot{V}*C_{H_{2}CO_{3}}-k*C_{H_{2}CO_{3}}*V)*(40mins-0mins) = V*(C_{H_{2}CO_{3},f}-{C_{H_{2}CO_{3}},i})$$
    
    
</div>
</details>
