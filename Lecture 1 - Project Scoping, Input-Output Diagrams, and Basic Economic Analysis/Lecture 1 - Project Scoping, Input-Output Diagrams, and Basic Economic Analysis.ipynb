<h2> Phases in the Construction of a Plant </h2>

Chemical plant construction can be split into four broad stages which come one after another:

<ol> 
    <li><strong> Basic Engineering</strong> - the scale of the plant, the draft of process flow diagram (PFD, we will introduce this later in the course), cost estimation, preliminary schedule, etc.</li>
    <br>
    <li> Detailed Engineering - client approved PFDs, detailed consturction schedule, real data of equipment costs from venders, etc.</li>
    <br>
    <li> Construction </li>
    <br>
    <li> Commissioning (start-up)</li>
    </ol>
    
<em> In this course, we will only focus on <strong>basic engineering </strong>.</em>

<h2> Process Design Scope </h2>

The discussion of what is inside and outside battery limits is usually established early in the design process to give a clear indication of what engineers are responsible for in their design.

<ul>
    <li><strong> Inside Battery Limits</strong>:  equipment and components that are associated with the main process feed streams [1], such as piping on process fluids.</li>
    <br>
    <li><strong> Outside Battery Limits</strong>: equipment, components, utilities, and other facilities acting on the utilities or waste treatment (not included in inside battery limits) that are associated with the process, but not related to the current design scope. [1].</li>
</ul>
<br>

<h2> Basic Process Economics </h2>

As with any venture, we want to ensure a chemical plant is profitable to operate. In order to get an initial sense of this in a project, we may perform a basic economic analysis. This analysis relies on four basic figured of merit and we will see how these are used to perform a basic economic analysis on a process.

The four basic figures of merit are:

<ol>
    <li> <strong>Gross Economic Potential (GEP) </strong>: generally in dollars/yr</li>
    GEP = value of products - value of feeds <br>
        <ul>
            <li> GEP has to be positive, otherwise our product is worth less than our feed and we will definitely be loosing money accounting for the operating costs. </li>
        </ul>
    <br>
    <li> <strong>Net Economic Potential (NEP) </strong>: generally in dollars/yr</li>
    NEP = GEP - production costs (utilities, labour, maintainance) <br>
        <ul>
            <li> NEP again has to be positive, otherwise we cannot pay for the process to operate. </li>
        </ul>
    <br>
    <li> <strong> Capital Costs (CC) </strong>: generally in dollars</li>
    CC = cost of plant design and construction <br>
    <br>
    <li> <strong> Return on Investment (ROI) </strong>: generally in %yr </li>
    ROI = $\frac{NEP}{CC}$*100%  <br> 
        <ul>
            <li> ROI also has to be positive, generally most processes require a certain minimum ROI, typically above 10%. ROI represents the percentage of capital costs recovered each year. The higher the ROI is, the more favorable the process is.</li>
        </ul>
</ol>

One thing that is essential to find our first figure of merit, GEP, is chemical pricing data. We will look at how to find this data now as it is so essential for evaluating processes. For now, we will not focus on how to assess production and capital costs, but will revisit these later in the course.
        
<h3> Finding Chemical Pricing Data </h3>

 Finding this data can be very challenging. Companies typically want to keep pricing information secret (for competition reasons). The best way I have found to search for chemical pricing data is as follows:


<ol>
    <li> Go to <a href="https://tamu.libguides.com/chemicalengineering/cpd"> Texas A&M Chemical Pricing Database </a> </li>
    <li> Search the chemical by name (i.e. Ethylene) or CAS number (Chemical Abstracts Services; can be found on Wikipedia) </li>
    <li> Find a citation from 'ICIS Chemical Business' and note the edition and page number (i.e. 2018, April 6; 19)</li>
    <li> Go to <a href="http://search.library.ubc.ca/">UBC Library</a> and type in "ICIS Chemical Business" into the search bar and search</li>
    <li> Click on the source by Factiva, Business Source Ultimate </li>
    <li> Click on 'Browse Journal'</li>
    <li> Find the citation you found from the Texas A&M Database citation using the dates </li>
    <li> Once you find your article, click 'UBC eLink'. This should take you to a new page where you can access the full text </li>
    <li> Search through the article for the chemical of interest's price </li>
    
</ol>

This process is tedious and often takes some time, even for people experienced in this process. So don't get discouraged when looking for this data and feel free to ask questions (through the discussion board, in tutorials or in office hours, etc.).

<h3> Example: Production of Vinyl Chloride </h3>

Our client wants to produce 360,000 tonnes/yr (800 million lbs/yr) of vinyl chloride monomer. We decide to use the following reaction pathway, which contains two reaction steps as follows:
    
$$C_{2}H_{4} + Cl_{2} → C_{2}H_{4}Cl_{2} (1)$$
<br>
$$C_{2}H_{4}Cl_{2} → C_{2}H_{3}Cl + HCl (2) $$
    
    
We gather the following information on the process:

    <thead>
        <tr><th style="text-align:center">Name</th><th style="text-align:center">Formula</th><th style="text-align:center">Molecular Weight (g/mol)</th><th style="text-align:center">Price (USD/tonne)</th></tr>
    </thead>
    <tbody>
        <tr><td style="text-align:center">Ethene</td><td style="text-align:center">$C_{2}H_{4}$</td><td style="text-align:center">28.05</td><td style="text-align:center">733</td></tr>
        <tr><td style="text-align:center">Chlorine</td><td style="text-align:center">$Cl_{2}$</td><td style="text-align:center">70.91</td><td style="text-align:center">240</td></tr>
        <tr><td style="text-align:center">Dichloroethane</td><td style="text-align:center">$C_{2}H_{4}Cl_{2}$</td><td style="text-align:center">98.96</td><td style="text-align:center">300</td></tr>
        <tr><td style="text-align:center">Hydrogen Chloride</td><td style="text-align:center">$HCl$</td><td style="text-align:center">36.46</td><td style="text-align:center">140</td></tr>
        <tr><td style="text-align:center">Vinyl Chloride</td><td style="text-align:center">$C_{2}H_{3}Cl$</td><td style="text-align:center">62.50</td><td style="text-align:center">750</td></tr>
    </tbody>
    </table>

Assume as well that our production costs are $200\frac{USD}{tonne}$ of VC produced, and assume our VC production plant costs $50,000,000 USD$ to build. Note these values usually depend on the size of the plant, location, and other factors.
    
Calculate or find all four economic figures of merit.
            
<h3> Input-Output Diagram </h3>
    
We know that one of the product streams will be 360,000 tonnes/yr of vinyl chloride. Hydrogen chloride is produced in a 1:1 molar ratio with vinyl chloride:
    
$VC = 360,000 tonnes/yr*\frac{tonne-mol}{62.50 tonnes} = 5,760\frac{tonne-mol}{yr}$
    
$HCl = 5,760\frac{tonne-mol}{yr}*36.46\frac{tonnes}{tonne-mol}=210,000\frac{tonnes}{yr}$
    
Similarly for ethane and chlorine, we get:
    
$C_{2}H_{4} = 5,760\frac{tonne-mol}{yr}*28.05\frac{tonnes}{tonne-mol}=162,000\frac{tonnes}{yr}$ 
    
$Cl_{2} = 5,760\frac{tonne-mol}{yr}*70.91\frac{tonnes}{tonne-mol}=408,000\frac{tonnes}{yr}$    
<br>

<h3> Process Economics </h3>
    
1 - Gross Economic Potential (GEP) in USD/yr
    
<p style="text-align:center;"><strong> GEP = Value of Products - Value of Feeds </strong></p>
    
$$GEP=140\frac{USD}{tonne}*210,000\frac{tonnes}{yr} + 750\frac{USD}{tonne}*360,000\frac{tonnes}{yr}-733\frac{USD}{tonne}*162,000\frac{tonnes}{yr}-240\frac{USD}{tonne}*408,000\frac{tonnes}{yr}$$
    
$$GEP=83,000,000\frac{USD}{tonne}$$
    
    
2 - Net Economic Potential (NEP) in USD/yr
    
<p style="text-align:center;"><strong> NEP = GEP - production costs</strong> </p> 
    
Using our assumed production costs of $200\frac{USD}{tonne}$ of VC produced.
    
$$NEP=83,000,000\frac{USD}{tonne}-200\frac{USD}{tonne}*360,000\frac{tonne}{yr}$$
    
$$NEP=11,000,000\frac{USD}{yr}$$
    

3 - Capital Costs (CC) in USD
    
<p style="text-align:center;"><strong> Total cost of plant design and construction</strong> </p>
    
This will be our VC production plant capital building cost of $50,000,000 USD$. 
    

4 - Return on Investment (ROI) in %/yr
    
<p style="text-align:center;"><strong> ROI = NEP/CC*100% </strong></p>   
    
$$ROI=11,000,000\frac{USD}{yr}*\frac{1}{50,000,000 USD}*100%$$
    
$$ROI=22 \frac{\%}{yr}$$ 
    
ROI is generally written simply in % and this is assumed to be %/yr. The higher the ROI the better.
